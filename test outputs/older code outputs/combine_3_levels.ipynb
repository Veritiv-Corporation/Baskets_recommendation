```python
import pandas as pd
# Load items 
df = pd.read_excel('digital_commerce_items.xlsx', usecols=['Item Number', 'Item Description'])
print(f">> Loaded {len(df)} e-commerce items and their descriptions")
```

```python
df.rename(columns={'Item Number': 'item_cde','Item Description':'description'}, 
          inplace=True)
print(f">> Renamed the columns to {df.columns.tolist()}")
```

```python
item_description_map = dict(zip(df['item_cde'], df['description']))
```

```python
product_result_100_freq=pd.read_csv('result_with_freq_cat_100_10.csv')
```

```python
# Define a function to convert floating-point numbers to strings and remove the .0 part
def remove_decimal(value):
    if pd.notna(value):
        return (int(value))
    else:
        return ''

# Assuming product_result_100_freq is your DataFrame

# Convert columns Recommendation 2 through Recommendation 5 to strings and remove the .0 part
product_result_100_freq['Recommendation 2'] = product_result_100_freq['Recommendation 2'].apply(remove_decimal)
product_result_100_freq['Recommendation 3'] = product_result_100_freq['Recommendation 3'].apply(remove_decimal)
product_result_100_freq['Recommendation 4'] = product_result_100_freq['Recommendation 4'].apply(remove_decimal)
product_result_100_freq['Recommendation 5'] = product_result_100_freq['Recommendation 5'].apply(remove_decimal)

```

```python
cat3_result=pd.read_csv('cat3_level_item_cde_to_reco_2.csv')
```

```python
# Convert columns Recommendation 2 through Recommendation 5 to strings and remove the .0 part
cat3_result['recommended_item_cde_2'] = cat3_result['recommended_item_cde_2'].apply(remove_decimal)
cat3_result['recommended_item_cde_3'] = cat3_result['recommended_item_cde_3'].apply(remove_decimal)
cat3_result['recommended_item_cde_4'] = cat3_result['recommended_item_cde_4'].apply(remove_decimal)
cat3_result['recommended_item_cde_5'] = cat3_result['recommended_item_cde_5'].apply(remove_decimal)
```

```python
for index, row in cat3_result.iterrows():
    print(row)
    for i in range(1, 6):
        col_name = f"recommended_item_cde_{i}"
        description_col_name = f"Description {i}"
        recommended_item_cde = row[col_name]
        if recommended_item_cde in item_description_map:
            cat3_result.at[index, description_col_name] = item_description_map[recommended_item_cde]

# Display the updated cat3_result
#print(cat3_result)

```

```python
def map_description(row):
    # Map 'item_cde' to 'description', if available
    description_1 = item_description_map.get(row['recommended_item_cde_1'], '')
    description_2 = item_description_map.get(row['recommended_item_cde_2'], '')
    description_3 = item_description_map.get(row['recommended_item_cde_3'], '')
    description_4 = item_description_map.get(row['recommended_item_cde_4'], '')
    description_5 = item_description_map.get(row['recommended_item_cde_5'], '')
    
    # Return a Series with the mapped descriptions
    return pd.Series([description_1, description_2, description_3, description_4, description_5])

# Apply the function to create new columns 'description 1' through 'description 5'
cat3_result[['description 1', 'description 2', 'description 3', 'description 4', 'description 5']] = cat3_result.apply(map_description, axis=1)
```

```python

# Assuming df is your original DataFrame with columns item_cde and description
# Assuming product_result_100_freq, cat3_result, and cat1_result are your DataFrames with recommendations

# Rename columns of cat3_result and cat1_result to match product_result_100_freq
cat3_result = cat3_result.rename(columns={'recommended_item_cde_1': 'Recommendation 1',
                                          'recommended_item_cde_2': 'Recommendation 2',
                                          'recommended_item_cde_3': 'Recommendation 3',
                                          'recommended_item_cde_4': 'Recommendation 4',
                                          'recommended_item_cde_5': 'Recommendation 5',
                                          'description 1': 'Description 1',
                                          'description 2': 'Description 2',
                                          'description 3': 'Description 3',
                                          'description 4': 'Description 4',
                                          'description 5': 'Description 5'})

cat1_result = cat1_result.rename(columns={'recommended_item_cde_1': 'Recommendation 1',
                                          'recommended_item_cde_2': 'Recommendation 2',
                                          'recommended_item_cde_3': 'Recommendation 3',
                                          'recommended_item_cde_4': 'Recommendation 4',
                                          'recommended_item_cde_5': 'Recommendation 5',
                                          'description 1': 'Description 1',
                                          'description 2': 'Description 2',
                                          'description 3': 'Description 3',
                                          'description 4': 'Description 4',
                                          'description 5': 'Description 5'})

```

```python
df_merged_product = pd.merge(df, product_result_100_freq[['item_cde', 'Recommendation 1', 'Description 1', 
                                                           'Recommendation 2', 'Description 2', 
                                                           'Recommendation 3', 'Description 3', 
                                                           'Recommendation 4', 'Description 4', 
                                                           'Recommendation 5', 'Description 5', 
                                                           'Level']], on='item_cde', how='left')
```

```python
# Iterate through each row in df_merged_product
for index, row in df_merged_product.iterrows():
    item_cde = row['item_cde']
    # Check if Recommendation columns are NaN
    if pd.isna(row['Recommendation 1']):
        # Find the corresponding row index in cat3_result
        cat3_index = cat3_result.index[cat3_result['item_cde'] == item_cde]
        if not cat3_index.empty:
            cat3_index = cat3_index[0]  # Extract the first index if multiple matches found
            # Update the values in df_merged_product with values from cat3_result
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Recommendation 1')] = cat3_result.iloc[cat3_index]['Recommendation 1']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Description 1')] = cat3_result.iloc[cat3_index]['Description 1']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Recommendation 2')] = cat3_result.iloc[cat3_index]['Recommendation 2']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Description 2')] = cat3_result.iloc[cat3_index]['Description 2']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Recommendation 3')] = cat3_result.iloc[cat3_index]['Recommendation 3']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Description 3')] = cat3_result.iloc[cat3_index]['Description 3']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Recommendation 4')] = cat3_result.iloc[cat3_index]['Recommendation 4']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Description 4')] = cat3_result.iloc[cat3_index]['Description 4']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Recommendation 5')] = cat3_result.iloc[cat3_index]['Recommendation 5']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Description 5')] = cat3_result.iloc[cat3_index]['Description 5']
            df_merged_product.iloc[index, df_merged_product.columns.get_loc('Level')] = cat3_result.iloc[cat3_index]['Level']

# Display the updated DataFrame
print(df_merged_product)
```
