```python
import pandas as pd
df_1=pd.read_csv("data_ecom_with_recos_1.csv")
df_2=pd.read_excel("top_5_items_catwise.xlsx")
```

```python
df_1.head()
```

```python
cond_for_same= df_1['reco1'].isnull() & df_1['reco5'].isnull()
df_1[cond_for_same]
```

# fill no reco values with it's own cat (upsell)

```python
#df_1['reco5'].fillna(df_1['Category'], inplace=True)

df_1.loc[cond_for_same, 'Level'] = 'Cat 3 (same)'
```

```python
# fill now empty level (cat3) coz other two are filled
df_1['Level'].fillna('Cat 3', inplace=True)
```

```python
df_1.loc[cond_for_same, 'reco5'] = df_1.loc[cond_for_same, 'Category']
```

```python
df_2.head()
```

```python
import ast

df_2['item_cde'] = df_2['item_cde'].apply(ast.literal_eval)
```

```python
df_2['item_cde'][2][2]
```

```python
condition = df_1['Level'] == 'Cat 3 (same)'
df_1[condition]
```

```python
mapping_dict_5 = df_2.set_index('cat')['item_cde'].apply(lambda x: x[0]).to_dict()

df_1.loc[condition, 'reco5'] = df_1.loc[condition, 'Category'].map(mapping_dict_5)
```

```python

mapping_dict_4 = df_2.set_index('cat')['item_cde'].apply(lambda x: x[1] if len(x) > 1 else None).to_dict()
df_1.loc[condition, 'reco4'] = df_1.loc[condition, 'Category'].map(mapping_dict_4)

mapping_dict_3 = df_2.set_index('cat')['item_cde'].apply(lambda x: x[2] if len(x) > 2 else None).to_dict()
df_1.loc[condition, 'reco3'] = df_1.loc[condition, 'Category'].map(mapping_dict_3)

mapping_dict_2 = df_2.set_index('cat')['item_cde'].apply(lambda x: x[1] if len(x) > 3 else None).to_dict()
df_1.loc[condition, 'reco2'] = df_1.loc[condition, 'Category'].map(mapping_dict_2)

mapping_dict_1 = df_2.set_index('cat')['item_cde'].apply(lambda x: x[1] if len(x) > 4 else None).to_dict()
df_1.loc[condition, 'reco1'] = df_1.loc[condition, 'Category'].map(mapping_dict_1)
```

```python
df_1.to_csv('only_same_cat_done.csv')
```

```python
mapping_dict = df_2.set_index('cat')['item_cde'].to_dict()
no_prod_cond= df_1[df_1['Level']=='Cat 3']
# Replace values in the 'reco1' column of df_1 with corresponding 'item_cde' values from df_2
df_1.loc[no_prod_cond, 'reco1'] = df_1.loc[no_prod_cond, 'reco1'].map(mapping_dict)
df_1.loc[no_prod_cond, 'reco2'] = df_1.loc[no_prod_cond, 'reco2'].map(mapping_dict)
df_1.loc[no_prod_cond, 'reco3'] = df_1.loc[no_prod_cond, 'reco3'].map(mapping_dict)

df_1.loc[no_prod_cond, 'reco4'] = df_1.loc[no_prod_cond, 'reco4'].map(mapping_dict)
df_1.loc[no_prod_cond, 'reco5'] = df_1.loc[no_prod_cond, 'reco5'].map(mapping_dict)
```

```python
data_ecom_cat_1=pd.read_csv('data_ecom_cat_1.csv')
```

```python
filtered_df_1 = df_1[df_1['Level'] == 'Cat 3 (same)']

# Merge filtered_df_1 with data_ecom_cat_1 based on the 'category' column
merged_df = pd.merge(filtered_df_1, data_ecom_cat_1, on='Category', how='left', suffixes=('_df1', '_data_ecom_cat_1'))

# Update 'reco1', 'reco2', 'reco3', 'reco4', 'reco5' columns in df_1 with corresponding values from data_ecom_cat_1
df_1.loc[filtered_df_1.index, ['reco1', 'reco2', 'reco3', 'reco4', 'reco5']] = merged_df[['reco1_data_ecom_cat_1', 'reco2_data_ecom_cat_1', 'reco3_data_ecom_cat_1', 'reco4_data_ecom_cat_1', 'reco5_data_ecom_cat_1']]
```
