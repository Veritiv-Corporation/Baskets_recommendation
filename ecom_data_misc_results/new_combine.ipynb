```python
def swap_pairs(row):
    if row['Level'] != 'Product':
        # Define pairs of columns and corresponding titles
        pairs = [(1, 5), (2, 4)]  # Define the pairs you want to swap
        
        for pair in pairs:
            reco_col1 = f'reco{pair[0]}'
            title_col1 = f'title{pair[0]}'
            reco_col2 = f'reco{pair[1]}'
            title_col2 = f'title{pair[1]}'
            
            # Swap values of reco columns
            row[reco_col1], row[reco_col2] = row[reco_col2], row[reco_col1]
            
            # Swap values of title columns
            row[title_col1], row[title_col2] = row[title_col2], row[title_col1]
            
    return row

# Apply the function to each row
df_1 = df_1.apply(swap_pairs, axis=1)
```
